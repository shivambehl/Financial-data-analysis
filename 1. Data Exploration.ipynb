### Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
##### {Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def averagePrice(N):
    df = idbi_df.tail(N)
    average_price = df['Close Price'].sum()/N
    return average_price
```

```python
def profitLossPercentage(N):
    df = idbi_df.tail(N)
    average_profit_loss_percentage = (df['Close Price'].iloc[-1] - df['Close Price'].iloc[0])*100/df['Close Price'].iloc[0]
    return average_profit_loss_percentage
```

```python
print('Average Price          ', averagePrice(7), '\nProfit Loss Percentage ', profitLossPercentage(7))
print('Average Price          ', averagePrice(14), '\nProfit Loss Percentage ', profitLossPercentage(14))
print('Average Price          ', averagePrice(30), '\nProfit Loss Percentage ', profitLossPercentage(30))
print('Average Price          ', averagePrice(90), '\nProfit Loss Percentage ', profitLossPercentage(90))
print('Average Price          ', averagePrice(180), '\nProfit Loss Percentage ', profitLossPercentage(180))
print('Average Price          ', averagePrice(365), '\nProfit Loss Percentage ', profitLossPercentage(365))
```

### Query 1.7
Add another column 'Trend' whose values are:

###### 'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
###### 'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
###### 'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
###### 'Positive' for 'Day_Perc_Change' in between 1 and 3
###### 'Negative' for 'Day_Perc_Change' in between -1 and -3
###### 'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
###### 'Among top losers' for 'Day_Perc_Change' in between -3 and -7
###### 'Bull run' for 'Day_Perc_Change' >7
###### 'Bear drop' for 'Day_Perc_Change' <-7

```python
idbi_df = idbi
```

```python
def trendSetter(df):
    if -0.5<df<= 0.5:
        return 'Slight or No Change'
    elif 0.5<df<=1:
        return 'Slight Positive'
    elif -1<df<=-0.5:
        return 'Slight Negative'
    elif 1<df<=3:
        return 'Positive'
    elif -3<df<=-1:
        return 'Negative'
    elif 3<df<=7:
        return 'Among Top Gainers'
    elif -7<df<=-3:
        return 'Among Top Losers'
    elif df>7:
        return 'Bull Run'
    elif df<-7:
        return 'Bear Drop'
```

```python
idbi_df['Trend'] = idbi_df.Day_Perc_Change.apply(trendSetter)
```
